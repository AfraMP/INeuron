```python
# Question 1 : Replacing NaN values in column 'FlightNumber' and make the column of type integer 

import pandas as pd
import numpy as np
df = pd.DataFrame({'From_To':['LoNDon_paris','MAdrid_miLAN','londON_StockhOlm','Budapest_PaRis','Brussels_londOn'],
                   'FlightNumber':[10045,np.nan,10065,np.nan,10085],
                   'RecentDelays':[[23, 47], [], [24, 43, 87], [13], [67, 32]],
                  'Airline':['KLM(!)', '<Air France> (12)', '(British Airways. )', '12. Air France', "Swiss Air"]})
df
```

```python
flightNumNanVal = np.where(df['FlightNumber'].isnull()==True)[0]
for i in flightNumNanVal:
    df.iloc[i,[1]] = df.iloc[i - 1, [1]] + 10
df['FlightNumber'] = df['FlightNumber'].astype('int64')
df
```

```python
# Question 2 : splitting From_To column into 'From' and 'To' column and assigning it to a temporary dataframe

df1 = pd.DataFrame()
df1['From']=''
df1['To']=''
splitFromTo = df['From_To'].apply(lambda x: x.split('_'))
for i in range(len(splitFromTo)):
    df1.loc[i, ['From']] = splitFromTo[i][0]
    df1.loc[i, ['To']] = splitFromTo[i][1]
df1
```

```python
# Question 3 : Standardizing the strings

header = df1.columns
for i in header:
    df1[i] = df1[i].apply(lambda x: x.title())
df1
```

```python
# Question 4 : Deleting From_To column and attaching temporary dataframe to df

df = df.drop('From_To', axis=1)
df = pd.concat([df1, df], axis=1)
df
```

```python
# Question 5 : Each 'Recentdelays' column values are separated into different columns

delays = pd.DataFrame()
delays['delay_1'] = ''
delays['delay_2'] = ''
delays['delay_3'] = ''
count = 0
header = delays.columns
noOfDelays = max(df['RecentDelays'].apply(lambda x: len(x)))
for i in range(len(df['RecentDelays'])):
    count = len(df['RecentDelays'][i])
    if count == 0:
        delays.loc[i, header[0]] =np.nan
    else:
        for x in range(count):
            delays.loc[i, [header[x]]] = df['RecentDelays'][i][x]

delays
```

```python
df = df.drop('RecentDelays', axis=1)
df.insert(3, 'delay_1', delays['delay_1'])
df.insert(4, 'delay_2', delays['delay_2'])
df.insert(5, 'delay_3', delays['delay_3'])
df
```
