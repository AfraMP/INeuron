```python
# Question 1 : Use Scipy to fit it to periodic function and plot the fit

from scipy import optimize
import matplotlib.pyplot as plt
import numpy as np
months=np.arange(12)
Max =[39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25]
Min = [21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18]

# Fitting a cosine function to the data

def temp(t, a,am, ti):
    return (a+am*np.cos((t-ti)*2*np.pi/t.max()))

res_max, cov_max = optimize.curve_fit(temp, months, Max)
res_min, cov_min = optimize.curve_fit(temp, months, Min)

# Plotting the resulting curve of data

month = np.linspace(0,12, num=12)
plt.figure()
plt.plot(months, Max,'ro')
plt.plot(month, temp(months, *res_max),'r-')
plt.plot(months, Min,'bo')
plt.plot(month, temp(months, *res_min),'b-')
plt.xlabel('Month')
plt.ylabel('Temperature (C)')
```

```python
# Question 2 : visualizaton using matplotlib
# 1. Pie chart presenting male/female proportion 

import pandas as pd
url = 'https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
titanic = pd.read_csv(url)
sizes = titanic['sex'].value_counts() 
label = ['Male', 'Female']
explode = [0, 0.1]
plt.pie(sizes, autopct='%.2f%%', labels=label, explode=explode, shadow=True)
```

```python
# 2. scatterplot with fare paid and age

color = {'female':0,'male':1}
sc = plt.scatter(titanic['fare'], titanic['age'], c=titanic['sex'].map(color))
cb = plt.colorbar(sc)
cb.set_ticks([0, 1])
cb.set_ticklabels(['female', 'male'])
plt.xlabel('Fare')
plt.ylabel('Age')
plt.title('Titanic Dataset')
```
