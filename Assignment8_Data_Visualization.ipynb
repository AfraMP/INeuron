```python
# Question 1 : Transform iris data into 3 dimensions and plot 3D chart 

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
```

```python
a = datasets.load_iris()
target = a.target
data = a.data
fig = plt.figure(figsize=(10,10))
ax = Axes3D(fig, elev=28)
# transforming iris data into 3 dimension using PCA
data_reduced = decomposition.PCA(n_components=3).fit_transform(data)
a = ax.scatter(data_reduced[:,0],data_reduced[:,1], zs=data_reduced[:,2], c = target,edgecolor='b',s=50)
cb = plt.colorbar(a)
cb.set_ticks([0, 1, 2])
cb.set_ticklabels(['setosa', 'versicolor', 'virginica'])
ax.set_title('Iris data')
ax.set_xlabel('1st eigenvector')
ax.set_ylabel('2nd eigenvector')
ax.set_zlabel('3rd eigenvector')
```
